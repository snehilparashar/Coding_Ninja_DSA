### Root to Node Path

```python
def rootTonodePath(root,target):
    
    if root is None:
        return None
    
    
    if root.data == target:
        l = list()
        l.append(root.data)
        return l
        
        
    leftlistoutput = rootTonodePath(root.left,target)
    
    if leftlistoutput != None:
        leftlistoutput.append(root.data)
        return leftlistoutput
    
    else: 
        return None
    
    rightlistoutput = rootTonodePath(root.right,target)
        
    if rightlistoutput != None:
        rightlistoutput.append(root.data)
        return rightlistoutput
    
    else:
        return None
```
